```python

# Convert SibSp, Parch, and Embarked to string type
df['SibSp'] = df['SibSp'].astype(str)
df['Parch'] = df['Parch'].astype(str)
df['Embarked'] = df['Embarked'].astype(str)
```

```python



fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(15, 12))
fig.suptitle('Countplot Visualization')

sns.countplot(x="Survived", data=df, hue='Sex', ax=axs[0][0])
axs[0][0].set_title('Survived With Respect To Sex')

sns.countplot(x="Survived", data=df, hue='SibSp', ax=axs[0][1])
axs[0][1].set_title('Survived With Respect To No Of Siblings')
axs[0][1].legend(title='SIBSP', loc='upper right')

sns.countplot(x="Survived", data=df, hue='Parch', ax=axs[1][0])
axs[1][0].set_title('Survived With Respect To No Of Parents & Children')
axs[1][0].legend(title='PARCH', loc='upper right')

sns.countplot(x="Survived", data=df, hue='Embarked', ax=axs[1][1])
axs[1][1].set_title('Survived With Respect To Embarked')
axs[1][1].legend(title='EMBARKED', loc='upper right')

plt.show()
```
